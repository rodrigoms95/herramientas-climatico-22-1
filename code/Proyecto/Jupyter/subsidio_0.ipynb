```python
# Calcula el pago de luz.

import pandas as pd
import numpy as np
```

```python
path = "../../../datos/CFE/"
fname = [ "Electrico_residencial_2010_2017.csv",
    "Tarifa_1.csv" ]

Consumo = ["Consumo"]
Consumo.append(
    [f"Consumo {x}" for x in range(2010, 2017)] )
Usuarios = ["Usuarios"]
Usuarios.append(
    [f"Usuarios {x}" for x in range(2010, 2017)] )

df = pd.read_csv(path + fname[0])

for i in [Consumo, Usuarios]:
    df[i[0]] = df[i[1]].sum(axis = 1)
    df.drop(i[1], axis = 1, inplace = True)

df["Consumo_Usuario"] = df["Consumo"] / df["Usuarios"]

df_DAC = df[df["Tarifa"] == "DAC"].copy()

df = df.where(df["Tarifa"] != "DAC").dropna()
df.reset_index(inplace = True)
df.drop("index", axis = 1, inplace = True)

tarifa_1 = pd.read_csv(path + fname[1])

escalon = tarifa_1["Escalon"].unique()

df[escalon] = np.nan

escalon
```

```python
df_dummy = df.copy()

for j in range(len(escalon) - 1):
    for i in df.itertuples():
        df_dummy.iloc[i[0], j + 9] = ( 
            tarifa_1.query('Temporada=="Verano"'
            ' & Escalon=="' + escalon[j] + '" & Tarifa=="' +
            str(i.Tarifa) + '"').iat[0,4] * 12 )

    df.iloc[:, j + 9] = df_dummy.iloc[:, j + 9]

for i in df.itertuples():
    if ( i.Consumo_Usuario
        <= (i[10] + i[11] + i[12]) ):
        df_dummy.loc[i[0], escalon[3]] = 0
        if ( i.Consumo_Usuario
            <= (i[10] + i[11]) ):
            df_dummy.loc[i[0], escalon[2]] = 0
            if ( i.Consumo_Usuario
                <= i[10] ):
                df_dummy.loc[i[0], escalon[1]] = 0
                df_dummy.loc[i[0], escalon[0]] = (
                    i.Consumo_Usuario )
            else:
                df_dummy.loc[i[0], escalon[1]] = (
                    i.Consumo_Usuario -
                    i[10]
                    )
        else:
            df_dummy.loc[i[0], escalon[2]] = (
                i.Consumo_Usuario -
                (i[10] + i[11])
                )
    else:
        df_dummy.loc[i[0], escalon[3]] = (
            i.Consumo_Usuario -
            (i[10] + i[11] + i[12])
            )

df_dummy.sort_values("Excedente", ascending = False).head()

df = df_dummy.copy()
```

```python
for i in df.itertuples():
    for j in range(len(escalon)):
        df_dummy.loc[i[0], escalon[j]] = ( i[j + 10] *
            tarifa_1.query('Temporada=="Verano"'
            ' & Escalon=="' + escalon[j] + '" & Tarifa=="' +
            str(i.Tarifa) + '"').iat[0,3] )

df = df_dummy.copy()

df_dummy.sort_values("Excedente", ascending = False).head()
```
