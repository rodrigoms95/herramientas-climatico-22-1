```python
# Verifica los municipios sin datos.

import rioxarray

import pandas as pd
import numpy as np

import xarray as xr
import geopandas as gpd
```

```python
path_d = "../../../resultados/CHIRTS/"
path_mask = "../../../resultados/Municipios/municipios_mask.nc"
path_shp = "../../../datos/Mapas/Municipios/mun20gw.shp"

vars = ["Tmax", "Tmin", "Tmean"]

# Abrimos los archivos a utilizar.
ds = ( xr.open_dataset(path_d + vars[0]
    + "/mexico_" + vars[0] + ".1995.nc") )

# Cargamos los municipios.
gdf = gpd.read_file( path_shp )

# Obtenemos el contorno de los municipios.
gdf["boundary"] = gdf.boundary

# Se establece el datum de los datos.
ds = ds.rio.write_crs(gdf.crs)
```
