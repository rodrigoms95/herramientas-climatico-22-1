```python
import numpy as np
import pandas as pd
import xarray as xr
import geoviews as gv

import geoviews.feature as gf
```

```python
path = "datos/"

fname = "cru_ts4.03.1901.2018.tmn.dat.nc"

tmn_ds = xr.load_dataset(path + fname)

tmn_ds
```

```python
kdims = ["lon","lat"]
tmn_vdims = "tmn"

tmn = gv.Dataset(tmn_ds, kdims = kdims, vdims = tmn_vdims)

tmn_ds.close()
```

```python
path = "datos/"

fname = "cru_ts4.03.1901.2018.tmx.dat.nc"

tmx_ds = xr.load_dataset(path + fname)

tmx_ds
```

```python
kdims = ["lon","lat"]
tmx_vdims = "tmn"

tmx = gv.Dataset(tmx_ds, kdims = kdims, vdims = tmx_vdims)

tmx_ds.close()
```

```python
gv.extension('matplotlib')
gv.output(size = 150)

opts = dict(colorbar = True)
```

```python
tmx.to(gv.Image, kdims).opts(**opts) * gf.coastline
```
